```python
# load in the "starbucks.csv" dataset
# refer to how we read the cereal.csv dataset in the tutorial
import pandas as pd

starbucks = pd.read_csv("../data/starbucks.csv")
starbucks
```

```python
# select all rows with more than and including 400 calories
starbucks.loc[starbucks.calories >= 400]
```

```python
# create a new column containing the caffeine per calories of each drink
starbucks['Caff_per_calo'] = starbucks['caffeine'] / starbucks['calories']
starbucks
```

```python
# what is the average calorie across all items?
starbucks.calories.mean()
```

```python
# how many different categories of beverages are there?
len(starbucks.beverage.unique())
```

```python
# what is the average # calories for each beverage category?
dic_cal_bev = {}
for typ in starbucks.beverage.unique():
    dic_cal_bev[typ] = starbucks.loc[starbucks.beverage == typ].calories.mean()
print(dic_cal_bev)
```
